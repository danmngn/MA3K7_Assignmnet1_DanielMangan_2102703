```python
import random

def hat_numbers(hat):

    while len(hat) > 1:
        draw = random.sample(hat, 2)

        difference = max(draw)-min(draw)

        hat.append(difference)
        hat.remove(min(draw))
        hat.remove(max(draw))
            
    return hat[0]

hat5 = [1, 2, 3, 4, 5]

final = hat_numbers(hat5)
print("The final number drawn is:", final)
```

```python
import random
import matplotlib.pyplot as plt

def hat_numbers(n):
    store = []

    for _ in range(1000):
        hat = list(range(1, n+1))

        while len(hat) > 1:
            draw = random.sample(hat, 2)

            difference = max(draw)-min(draw)

            hat.append(difference)

            hat.remove(min(draw))
            hat.remove(max(draw))

        store.append(hat[0])

    return store

n=5
all_simulations = hat_numbers(n)

plt.hist(all_simulations, bins=range(0,n+2), align='left', edgecolor = "black")
plt.xlabel('Final Number Drawn')
plt.ylabel('Frequency')
plt.title('1000 Simulations for n=5')
plt.show()
```
