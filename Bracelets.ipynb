```python
import numpy as np

#Computing the Chain
def chain(first, second, length):
    chain = [first, second]
    for i in range(2, length):
        chain.append( (chain[-1] + chain[-2])%10 )
    return chain

#Does the Chain Loop?
def findLoop(first, second, brace):
    
    array = []
    
    for i in range(len(brace) - 1): #The minus 1 is because we are working in pairs of numbers. if there was a chain of 100 numbers, there would be 99 adjacent pairs
        
        pair = (brace[i+1], brace[i+2])
        array.append(pair)

        if pair == (first, second):
            break
            
    return array

#Chain 1
Chain1 = chain(1, 5, 100) #We use 100, as that is the maximimum possibel length the chain could be

Bracelet1 = findLoop(1, 5, Chain1)
print("Bracelet 1:")
print()
print(Bracelet1)
print()
print("Length of Bracelet 1 is", len(Bracelet1))

#Chain 2
Chain2 = chain(0, 0, 100) #We use 100, as that is the maximimum possibel length the chain could be

Bracelet2 = findLoop(0, 0, Chain2)
print()
print("Bracelet 2:")
print()
print(Bracelet2)
print()
print("Length of Bracelet 2 is", len(Bracelet2))

#Chain 3
Chain3 = chain(0, 2, 100) #We use 100, as that is the maximimum possibel length the chain could be

Bracelet3 = findLoop(0, 2, Chain3)
print()
print("Bracelet 3:")
print()
print(Bracelet3)
print()
print("Length of Bracelet 3 is", len(Bracelet3))

#Chain 4
Chain4 = chain(0, 5, 100) #We use 100, as that is the maximimum possibel length the chain could be

Bracelet4 = findLoop(0, 5, Chain4)
print()
print("Bracelet 4:")
print()
print(Bracelet4)
print()
print("Length of Bracelet 4 is", len(Bracelet4))

#Chain 5
Chain5 = chain(1, 3, 100) #We use 100, as that is the maximimum possibel length the chain could be

Bracelet5 = findLoop(1, 3, Chain5)
print()
print("Bracelet 5:")
print()
print(Bracelet5)
print()
print("Length of Bracelet 5 is", len(Bracelet5))

#Chain 6
Chain6 = chain(2, 6, 100) #We use 100, as that is the maximimum possibel length the chain could be

Bracelet6 = findLoop(2, 6, Chain6)
print()
print("Bracelet 6:")
print()
print(Bracelet6)
print()
print("Length of Bracelet 6 is", len(Bracelet6))

Total_Pairings = 100

Pairings_in_Bracelets = len(Bracelet1) + len(Bracelet2) + len(Bracelet3) + len(Bracelet4) + len(Bracelet5) + len(Bracelet6)

print()
print()
print("There were", Total_Pairings, "possible pairings.  The amount of pairings used in the six computated bracelets were",
      Pairings_in_Bracelets, "Therefore we have exhausted all possibilities and there are 6 Bracelets in total.")

```
