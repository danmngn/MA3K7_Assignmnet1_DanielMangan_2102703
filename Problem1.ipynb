```python
import random

class InitConditions:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
        self.collisions = 0

    def collision(self, p1, p2):
        if p1 == p2:
            return
        if (p1 == 'A' and p2 == 'B') or (p1 == 'B' and p2 == 'A'):
            self.a -= 1
            self.b -= 1
            self.c += 2
            self.collisions += 1
        elif (p1 == 'A' and p2 == 'C') or (p1 == 'C' and p2 == 'A'):
            self.a -= 1
            self.b += 2
            self.c -= 1
            self.collisions += 1
        elif (p1 == 'B' and p2 == 'C') or (p1 == 'C' and p2 == 'B'):
            self.a += 2
            self.b -= 1
            self.c -= 1
            self.collisions += 1

    def simulation(self):
        while self.collisions < 100:
            particles = ['A'] * self.a + ['B'] * self.b + ['C'] * self.c 
            if len(particles) == 2:  
                p1, p2 = particles
                self.collision(p1, p2)
            else:  
                p1, p2 = random.sample(particles, 2)
                self.collision(p1, p2)
        print("Number of collisions:", self.collisions)
        print("Remaining particles in sealed container A:", self.a, "B:", self.b, "C:", self.c)
        print("N_T =", self.a + self.b + self.c)
        print()

simulate = InitConditions(14, 14, 15)
simulate.simulation()
simulate = InitConditions(14, 15, 15)
simulate.simulation()
simulate = InitConditions(22, 22, 23)
simulate.simulation()
simulate = InitConditions(22, 23, 23)
simulate.simulation()
```
