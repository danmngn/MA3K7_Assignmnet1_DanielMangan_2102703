```python
def fibonacci(n):
    if n <= 0:
        return "Invalid input"
    elif n == 1:
        return 1
    elif n == 2:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)

print('The 25th fibonacci number is:', fibonacci(25))
```

```python
import matplotlib.pyplot as plt
import numpy as np

def Prob(N):
    return 2/3 - 2/3 * (-1/2)**N

# Generate values for N
N1 = np.arange(1, 25, 1)
N2 = np.arange(1, 10, 1)
N3 = np.arange(1, 9, 1)
# Calculate corresponding SN values
SN1 = Prob(N1)
SN2 = Prob(N2)

# Plotting
plt.plot(N1, SN1)
plt.xlabel('N')
plt.ylabel('Probability of landing on SN')
plt.title('Values of P(SN)')
plt.grid(True)
plt.show()

plt.plot(N2, np.abs(SN2 - 2/3))
plt.xlabel('N')
plt.ylabel('Distance of P(SN) from 2/3')
plt.title('Values of Distance of P(SN) from 2/3')
plt.grid(True)
plt.show()

for i in range(len(N2)-1):
    factor = np.abs(SN2[i+1] - 2/3) / np.abs(SN2[i] - 2/3)
    print(f"For N={N2[i]}, the factor is {factor}")
    
print('The distance between the probability of landing on SN and 2/3 drops by a factor of 1/2 for every increment you increase N by 1.')
```
